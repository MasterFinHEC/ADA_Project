### Gredient descent

We want to find the minimum square error between our approximation and real shape of data. For this we pick some numbers and compute the square errors so we have some starting points for the newton gradient descent. We use a newton gradient descent to find the optimal parameters. The function and the gradients:
$$f(x) = \frac{L}{1+e^{-k(x_i - x_0)}$$


```python
# Setting to compute first derivative
def J(j,ytrue,x):
    gl = 0
    gk = 0
    gx0 = 0
    for i in range(len(ytrue)):
        gl = gl + (j[0]/(1+np.exp(j[1]*(j[2]-x[i])))-ytrue[i])/(1+np.exp(j[1]*(j[2]-x[i])))

        gk = gk + (ytrue[i]-j[0]/(1+np.exp(j[1]*(j[2]-x[i]))))*(j[0]/((1+np.exp(j[1]*(j[2]-x[i])))**2))*(j[2]-x[i])*np.exp(j[1]*(j[2]-x[i]))
    
        gx0 = gx0 + (ytrue[i]-j[0]/(1+np.exp(j[1]*(j[2]-x[i]))))*(j[0]/(1+np.exp(j[1]*(j[2]-x[i])))**2)*j[1]*np.exp(j[1]*(j[2]-x[i]))
        
    gl = 2*gl
    gk = 2*gk
    gx0 = 2*gx0
    p=np.array([gl,gk,gx0], dtype=np.float128)
    return p
```

```python
# Setting gradient descent function
def Newton(w,country,ytrue,x): 
    count = 0
    Maxcount = 10000
    Delta = 1 # Start with a random delta that will decrease,hopefully.
    alpha = 0.0001 #alpha
    while count<Maxcount and Delta>10**(-14):
        p = J(w,ytrue,x)
        wOld = w
        w = w - alpha*p
        Delta = np.sum((w-wOld)**2)
        count += 1
    ypred = np.zeros((len(ytrue),1),dtype=np.float128)
    for i in range(len(ytrue)):
        ypred[i] = w[0]/(1+np.exp(w[1]*(w[2]-x[i])))
    return w, ypred
```

```python
def best(country):
    for i, date in enumerate(y.columns):
        if y.loc[country,date] != 0:
            ytrue = y.loc[country,y.columns[i-1]:y.columns[-2]]
            break
    
    ytrue = np.array(ytrue/X.loc[country,'Pop']*100)
    x = np.linspace(0,len(ytrue)-1,len(ytrue)) 
    L = np.linspace(0,1,101)
    K = np.linspace(0,0.3,31)
    X0 = np.linspace(0,100,101)
    
    lb = 0
    kb = 0
    x0b = 0
    lsOld = 10000
    for i, l in enumerate(L):
        for j, k in enumerate(K):
            for t, z in enumerate(X0):                
                ls = 0
                for q in range(len(ytrue)):
                    ls = ls + (ytrue[q]-l/(1+np.exp(k*(z-q))))**2  
                if (ls < lsOld):
                    lb = L[i]
                    kb = K[j]
                    x0b = X0[t]
                    lsOld = ls
                else:
                    pass
                
    j = np.array([lb,kb,x0b])
    w, ypred = Newton(j, country, ytrue, x)
    return w, ypred, ytrue, x
```

```python
# Splitting the data
RanInd = np.random.permutation(Data.index)
y_train_L = Lmax.loc[RanInd[0:40]]
y_test_L = Lmax.loc[RanInd[41:60]]
y_train_K = Ks.loc[RanInd[0:40]]
y_test_K = Ks.loc[RanInd[41:60]]
y_train_x0 = X0s.loc[RanInd[0:40]]
y_test_x0 = X0s.loc[RanInd[41:60]]
x_train_L = Data.loc[RanInd[0:40],['Gini','Dem','Pop','Child','RDI','Political Stability','GOV']]
x_test_L = Data.loc[RanInd[41:60],['Gini','Dem','Pop','Child','RDI','Political Stability','GOV']]
x_train_K = Data.loc[RanInd[0:40],col_data]
x_test_K = Data.loc[RanInd[41:60],col_data]
x_train_x0 = Data.loc[RanInd[0:40],['Distance']]
x_test_x0 = Data.loc[RanInd[41:60],['Distance']]



# Import Gaussian Process 
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.linear_model import LinearRegression

# Predicting the value
#L
LR = LinearRegression()
GPR = GaussianProcessRegressor()
LR.fit(x_train_L,y_train_L)
y_predict_L = LR.predict(x_test_L)

GPR.fit(x_train_L,y_train_L)
y_GPRpredict_L = GPR.predict(x_test_L)

#K
LR.fit(x_train_K,y_train_K)
y_predict_K = LR.predict(x_test_K)
GPR.fit(x_train_K,y_train_K)
y_GPRpredict_K = GPR.predict(x_test_K)

#x0
LR.fit(x_train_x0,y_train_x0)
y_predict_x0 = LR.predict(x_test_x0)
GPR.fit(x_train_x0,y_train_x0)
y_GPRpredict_x0 = GPR.predict(x_test_x0)
```

### Neural Network

```python
X1 = Data.values
Y1 = Lmax.values
Y2 = Ks.values
Y3 = X0s.values
# define base model
def baseline_model():
    # create model
    model = Sequential()
    model.add(Dense(24, input_dim=24, kernel_initializer='normal', activation='relu'))
    model.add(Dense(1, kernel_initializer='normal'))
    # Compile model
    model.compile(loss='mean_squared_error', optimizer='adam')
    return model

# evaluate model
estimator = KerasRegressor(build_fn=baseline_model, epochs=100, batch_size=5, verbose=0)
kfold = KFold(n_splits=5)
resultsL = cross_val_score(estimator, X1, Y1, cv=kfold)
resultsK = cross_val_score(estimator, X1, Y2, cv=kfold)
resultsX0 = cross_val_score(estimator, X1, Y3, cv=kfold)
print("Baseline L: %.2f (%.2f) MSE" % (resultsL.mean(), resultsL.std()))
print("Baseline K: %.2f (%.2f) MSE" % (resultsK.mean(), resultsK.std()))
print("Baseline x0: %.2f (%.2f) MSE" % (resultsX0.mean(), resultsX0.std()))
```

```python
# evaluate model with standardized dataset
estimators = []
estimators.append(('standardize', StandardScaler()))
estimators.append(('mlp', KerasRegressor(build_fn=baseline_model, epochs=50, batch_size=5, verbose=0)))
pipeline = Pipeline(estimators)
kfold = KFold(n_splits=5)
results2L = cross_val_score(pipeline, X1, Y1, cv=kfold)
results2K = cross_val_score(pipeline, X1, Y2, cv=kfold)
results2X0 = cross_val_score(pipeline, X1, Y3, cv=kfold)
print("Standardized L: %.2f (%.2f) MSE" % (results2L.mean(), results2L.std()))
print("Standardized K: %.2f (%.2f) MSE" % (results2K.mean(), results2K.std()))
print("Standardized X0: %.2f (%.2f) MSE" % (results2X0.mean(), results2X0.std()))
```

```python
def larger_model():
    # create model
    model = Sequential()
    model.add(Dense(24, input_dim=24, kernel_initializer='normal', activation='relu'))
    model.add(Dense(6, kernel_initializer='normal', activation='relu'))
    model.add(Dense(1, kernel_initializer='normal'))
    # Compile model
    model.compile(loss='mean_squared_error', optimizer='adam')
    return model
# evaluate model with standardized dataset
estimators = []
estimators.append(('standardize', StandardScaler()))
estimators.append(('mlp', KerasRegressor(build_fn=larger_model, epochs=50, batch_size=5, verbose=0)))
pipeline = Pipeline(estimators)
kfold = KFold(n_splits=5)
results3L = cross_val_score(pipeline, X1, Y1, cv=kfold)
results3K = cross_val_score(pipeline, X1, Y2, cv=kfold)
results3X0 = cross_val_score(pipeline, X1, Y3, cv=kfold)
print("Larger L: %.2f (%.2f) MSE" % (results3L.mean(), results3L.std()))
print("Larger K: %.2f (%.2f) MSE" % (results3K.mean(), results3K.std()))
print("Larger X0: %.2f (%.2f) MSE" % (results3X0.mean(), results3X0.std()))
```
